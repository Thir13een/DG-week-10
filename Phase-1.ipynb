## Customer Trends Over Time

```python
# Analyzing how customer attributes change over time
```

```python
data['contract_start_date'] = pd.to_datetime(data['contract_start_date'])
data.set_index('contract_start_date').resample('M')['new_customer_index'].count().plot(figsize=(10, 5))
plt.title('New Customers Over Time')
plt.show()
```

#### Observations:

1) **Stable Early Growth:** Consistent acquisition strategies drove steady growth from 1995-2005.
---
2) **Campaign & Market Influence:** Sharp spikes from 2005-2015 indicate that targeted campaigns or market events greatly boosted new customer sign-ups.
---
3) **Retention Challenges:** The rapid declines following peaks highlight the need for continuous and balanced acquisition efforts to maintain growth.
---
4) **Recent Decline:** The sharp drop after 2014 suggests a need to investigate potential strategic gaps or market shifts affecting customer acquisition.
---
Overall, a mix of sustained strategies and impactful campaigns is key, with attention needed to reverse the recent decline in new customers.

## Credit Card Ownership Over Time

```python
# Tracking credit card adoption over different months to identify any seasonality or promotional impacts.
```

```python
credit_card_avg = data.groupby(data['contract_start_date'].dt.to_period('M'))['credit_card_status'].mean()
credit_card_avg.plot(kind='area', figsize=(10, 5), alpha=0.4)
plt.title('Area Plot of Average Credit Card Status Over Time')
plt.xlabel('Month')
plt.ylabel('Average Credit Card Status')
plt.show()
```

#### Observations:

The decline in average credit card ownership over time, with volatility and sharp dips, indicates reduced popularity or relevance. This trend suggests the need to understand changing customer preferences, enhance credit card offerings, or pivot to products better aligned with current market demands.

### Customer Segmentation and Clustering

```python
# KMeans clustering to create customer segments
```

```python
from sklearn.cluster import KMeans
```

```python
X = data[['customer_age', 'household_income', 'customer_seniority_months']]
kmeans = KMeans(n_clusters=5, n_init=10, random_state=42).fit(X)  # Set n_init explicitly
data['segment'] = kmeans.labels_

sns.scatterplot(data=data, x='customer_age', y='household_income', hue='segment', palette='Set2')
plt.title('Customer Segments Based on Age and Income')
plt.legend(loc='upper right')  # Specify a fixed legend location
plt.show()
```

#### Observations:

1) The segmentation reveals clear customer clusters based on age, income, and seniority, allowing for more targeted strategies.
---
2) Age and household income play significant roles in distinguishing customer behavior, helping to identify specific needs and financial preferences across segments.
---
3) The well-separated clusters indicate potential to design personalized marketing and product offerings for different customer groups, enhancing customer engagement and product relevance.
---
4) Some overlap suggests that customers with similar income levels may share financial behaviors despite age differences, allowing for broader product applicability within those groups.

-----


Overall, the clustering analysis enables more focused targeting of financial products and personalized customer experiences.
